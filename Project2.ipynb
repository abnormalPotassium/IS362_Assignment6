## Tidying up Religion and Income Data

I chose Robert's post for my first dataset and converted his image into a csv. The current problem with this dataset is that there are three variables of religion, income, and frequency however there are more than 3 columns. The Income variable's values are being used as column headers. This is what we want to fix.

```python
df_relinc = df_relinc.melt(id_vars = ["religion"],var_name = ["income"],value_name = "frequency")
df_relinc.head(10)
```

## Analysis of the Tidied Data

Now that our data is tidied we would be able to better analyze the relationship between income and religion. As the data is now more friendly to manipulation, we would be able to display a graph if we wanted to or see the distribution of people who refused to tell about their religion.

```python
df_relinc[df_relinc['religion'] == 'refused'].sort_values('frequency', ascending = False).head(10)
```

## Tidying up UN Migrant Data
I chose Kenneth's post for my third dataset and converted a sheet from the 2015 migrant data into a csv. There are quite a few problems with this dataset that stem from the sheet being setup for excel readibility rather than analyzing with pandas. We want to rename columns that don't have any labels, filter out the introductory data such as regions near the top, take out all the columns that are unused, and then make it so origin countries are a column instead of headers.

```python
df_immigrants.rename(columns={'Unnamed: 3': 'Destination Country Code','Unnamed: 1': 'Destination Country'},inplace=True)

df_immigrants = df_immigrants[df_immigrants['Unnamed: 4'].notna()]

df_immigrants.drop(['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4', 'Total'],axis=1,inplace=True)

df_immigrants = df_immigrants.melt(id_vars=['Destination Country', 'Destination Country Code'], value_name='People', var_name='Origin Country')
```
